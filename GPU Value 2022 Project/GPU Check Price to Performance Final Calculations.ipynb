# Extra work for someone who wants 1440p info...

```python
GPU=pd.read_csv('GPU_Prices_And_Performance.csv',index_col=0)
GPU = GPU.drop(['NEWUSPrice','AVG1080pFPS','AVG4kFPS'],axis=1) 
GPU

# Dollars per frame, used
GPU['Dollars_per_frame_used'] = GPU['Ebay_Used_Price'] / GPU['AVG1440pFPS']
# Dollars per frame, new
GPU['Dollars_per_frame_new'] = GPU['Newegg_New_Price'] / GPU['AVG1440pFPS']
GPU
New_GPU = GPU[['GPU','AVG1440pFPS','Newegg_New_Price','Dollars_per_frame_new']]
New_GPU = New_GPU.sort_values('Dollars_per_frame_new',ascending=True)

Used_GPU = GPU[['GPU','AVG1440pFPS','Ebay_Used_Price','Dollars_per_frame_used']]
Used_GPU = Used_GPU.sort_values('Dollars_per_frame_used',ascending=True)

New_GPU
```

```python
Used_GPU
```

# The lowest acceptable performance for 1440p IMO is found in either the 6700XT or the 3070, and although the 3070 performs better by about 11 frames per second, it will cost 100 USD higher.

### If you're going 1440p, you're gonna want a GPU that doesn't make you want to turn the details down (or revert to 1080p) to have your fps acceptably high.
### If you have a 1440p 60fps monitor, you can probably just aim for 90-100 fps, ie: 3060/6600XT range. Aiming for a card with a 60fps average is not a good idea, most games have areas where the card has to work much harder for extended periods of time. A card like the 1660 will drop to 30-40 often.

# I would say you should avoid the used market for an above 400 dollar investment, but the 3070 ti at 450 dollars used is a pretty good price, if you have the PSU to back it up. (290w just for the GPU alone)
