# The goal of this project is to scrape for the best deals on Graphics cards across the internet, both New and Used (ebay). It is also a way to see at any point of time what cards give you the best bang for the buck.

```python
import requests
from bs4 import BeautifulSoup
```

```python
link='https://www.videocardbenchmark.net/high_end_gpus.html'
```

```python
html=requests.get(link).text
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
gpunames=list(soup.findAll('span',{'class':'prdname'}))
gpulist=[]
for entry in gpunames:
    new=str(entry)
    new=str.replace(new,'<span class="prdname">','')
    new=str.replace(new,'</span>','',)
    gpulist.append(new)
len(gpulist)
```

```python
newprices=list(soup.findAll('span',{'class':"price-neww"}))
newpricelist=[]
for entry in newprices:
    new=str(entry)
    new=str.replace(new,'<span class="price-neww">','')
    new=str.replace(new,'</span>','')
    new=str.replace(new,'*','')
    new=str.replace(new,'NA','np.nan')
    newpricelist.append(new)
len(newpricelist)

#checking to make sure the size of both lists are equal
```

```python
scores=list(soup.findAll('span',{'class' :'count'}))
scorelist=[]
for entry in scores:
    new=str(entry)
    new=str.replace(new,'<span class="count">','')
    new=str.replace(new,'</span>','')
    scorelist.append(new)
len(scorelist)

#checking to make sure the size of all lists are equal
```

```python
import pandas as pd
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
import numpy as np
df=pd.DataFrame([gpulist,newpricelist,scorelist]).T

df=df.reset_index()
df.rename(columns={'index':'RANKING',0:'GPU',1:'NEW_PRICE',2:'PASSMARK_SCORE'},inplace=True)
df['RANKING']=df['RANKING']+1
df=df.head(158) #removing irrelevant older GPUs
                #worst gpu in the list is the RX 580, for those interested.
```
