```python
temp=USED.sort_values('RANKING',ascending=True)
temp=temp.reset_index().drop('index',axis=1)
list_=[]
for i in range(0,45):
    if temp['NEW_PRICE'][i] >0:
        if temp['USED_PRICE'][i] >0:
            list_.append(i)
temp=temp.loc[list_,:]
(temp['NEW_PRICE']/temp['USED_PRICE']).mean()
```
